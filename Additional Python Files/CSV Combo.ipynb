## Traffic

```python
csv_files_traffic = glob.glob('Traffic/*.csv')
```

## Events

```python
import numpy as np
events = pd.read_csv('Events.csv')
sf_events = events[['Date of Event', 'State Farm Arena']]
mb_events = events[['Date of Event', 'MB Stadium']]
```

```python
sf_events['SF_Event']= np.where(sf_events['State Farm Arena'].notna() == True, 1, 0)
sf_events['dt'] = sf_events['Date of Event'] + " " + sf_events['State Farm Arena']
sf_events = sf_events.dropna()
sf_events = sf_events[['dt', 'SF_Event']]
sf_events['dt'] = pd.to_datetime(sf_events['dt'], format="%d-%b-%y %H:%M")
```

```python
mb_events['MB_Event'] = np.where(mb_events['MB Stadium'].notna() == True, 1, 0)
mb_events['dt'] = mb_events['Date of Event'] + ' ' + mb_events['MB Stadium']
mb_events= mb_events.dropna()
mb_events= mb_events[['dt', 'MB_Event']]
mb_events['dt'] = pd.to_datetime(mb_events['dt'], format="%d-%b-%y %H:%M")
```

```python
table = pd.read_csv('merged.csv')
table = table.drop(columns = 'Unnamed: 0', axis =1)
table = table.drop(columns='TRIP_ID')
table = table.dropna()
```

```python
table['last_updated'] = pd.to_datetime(table['last_updated'])
```

```python
table2 = table.merge(sf_events, how = 'left', left_on = 'last_updated', right_on = 'dt')
table3 = table2.merge(mb_events, how = 'left', left_on = 'last_updated', right_on = 'dt')
```

```python
data = table3.drop(columns = ['dt_y', 'dt_x'])
data['SF_Event'] = data['SF_Event'].fillna(0)
data['MB_Event'] = data['MB_Event'].fillna(0)
```
